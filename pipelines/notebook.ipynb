## Usage in VSCode
Jupyter Notebook has a nice integration with VSCode. When running VSCode, make sure that:
1. you installed the dependencies with `poetry install`
2. you started VSCode for the first time with `poetry run code .`
3. select the correct environment in the top-right of the Notebook window, so you work actually in the correct `poetry` environment.

Jupyter Notebook in VSCode

These action are only required once, as VSCOde will persist your project settings.

When runnning this notebook, VSCode will may complain that the `ipykernel` is not installed. In the dialog, you MUST select the option 'Change kernel' and select the correct Python environment. When the correct environment is not listed, please make sure that you started the IDE using `poetry` (step 2).
